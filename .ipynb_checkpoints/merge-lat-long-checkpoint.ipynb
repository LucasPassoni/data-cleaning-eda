```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#carregando a base
dados_2021 = pd.read_excel("seguro-rural-limpo-2021.xlsx", decimal=',')
dados_2021.info()
```

```python
dados_2021['apolice']
```

```python
#carregando a base com latitude e longitude
dados_2016_2021_lat_long = pd.read_excel("dados-2016-2021-latitude-longitude.xlsx", decimal=',')
```

```python
#transformando o tipo apolice
dados_2016_2021_lat_long['apolice'] = dados_2016_2021_lat_long['apolice'].astype(str)
dados_2016_2021_lat_long.info()
```

```python
dados_2016_2021_lat_long['apolice'] = dados_2016_2021_lat_long['apolice'].str.replace(".0", "")
dados_2016_2021_lat_long.head(10).sort_values(by='apolice', ascending=True)
```

```python
dados_2016_2021_lat_long['latitude'] = dados_2016_2021_lat_long['latitude'].str.replace("--", "-")
dados_2016_2021_lat_long['latitude']
```

```python
dados_2021_lat_long = dados_2021.merge(dados_2016_2021_lat_long, left_on='apolice', right_on='apolice', how='left')
```

```python
dados_2021_lat_long
```

```python
dados_2021_lat_long.to_excel('seguro-rural-limpo-2021-lat-long.xlsx')
```
