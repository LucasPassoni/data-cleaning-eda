```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.mode.copy_on_write = True
```

```python
dados_limpo_2021 = pd.read_excel("seguro-rural-limpo-2021.xlsx", decimal=',')
dados_limpo_2021.info()
```

```python
#carregando a base
dados_2021 = pd.read_excel("dados-2016-2021.xlsx", decimal=',')
dados_2021.info()
```

```python
#salvando os dados
dados_lat_long = dados_2021[['NR_APOLICE', 'LATITUDE', 'NR_GRAU_LAT', 'NR_MIN_LAT', 'NR_SEG_LAT', 
                            'LONGITUDE', 'NR_GRAU_LONG', 'NR_MIN_LONG', 'NR_SEG_LONG']]
dados_lat_long
```

```python
#alterando os nomes da coluna
dados_lat_long.columns = ['apolice', 'latitude', 'lat.grau', 'lat.min', 'lat.seg', 'longitude', 'long.grau', 'long.min', 'long.seg']
pd.DataFrame(dados_lat_long)
```

```python
#transformando os minutos e segundos em graus
dados_lat_long['lat.min'] = dados_lat_long['lat.min']/60
dados_lat_long['long.min'] = dados_lat_long['long.min']/60

dados_lat_long['lat.seg'] = dados_lat_long['lat.seg']/3600
dados_lat_long['long.seg'] = dados_lat_long['long.seg']/3600
```

```python
dados_lat_long
```

```python
#somando graus minutos e segundos
dados_lat_long['lat.grau'] = dados_lat_long['lat.grau']+dados_lat_long['lat.min']+dados_lat_long['lat.seg']
dados_lat_long['long.grau'] = dados_lat_long['long.grau']+dados_lat_long['long.min']+dados_lat_long['long.seg']
dados_lat_long
```

```python
#trocanso Sul e Oeste pelo sinal de negativo
dados_lat_long = dados_lat_long[['apolice', 'latitude', 'lat.grau', 'longitude', 'long.grau']]

dados_lat_long['latitude'] = dados_lat_long['latitude'].str.replace('S', '-')
dados_lat_long['longitude'] = dados_lat_long['longitude'].str.replace('W', '-')

dados_lat_long['lat.grau'] = dados_lat_long['lat.grau'].astype(str)
dados_lat_long['latitude'] = dados_lat_long[['latitude', 'lat.grau']].agg(" ".join, axis=1)

dados_lat_long['long.grau'] = dados_lat_long['long.grau'].astype(str)
dados_lat_long['longitude'] = dados_lat_long[['longitude', 'long.grau']].agg(" ".join, axis=1)
dados_lat_long
```

```python
dados_lat_long = dados_limpo_2021.merge(dados_lat_long, left_on='apolice', right_on='apolice', how='left')
```

```python
dados_lat_long
```

```python
dados_lat_long.to_excel('seguro-rural-limpo-2021-lat-long.xlsx')
```
